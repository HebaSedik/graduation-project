```python
from ultralytics import YOLO
import cv2
import math
import cvzone  # Make sure you have installed this library
model = YOLO('fire.pt')

cap = cv2.VideoCapture(0)  # Update the video source as needed

if not cap.isOpened():
    print("Failed to open the video capture.")
    exit()

classnames = ["fire"]  # Update with your class names

while True:
    ret, frame = cap.read()

    if not ret:
        print("Failed to read frame from the video capture.")
        break

    frame = cv2.resize(frame, (640, 480))
    result = model(frame, stream=True)  # Make sure you have defined and loaded the model

    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            Class = int(box.cls[0])
            if confidence > 50:
                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 5)
                cvzone.putTextRect(frame, f'{classnames[Class]} {confidence}%', [x1 + 8, y1 + 100],
                                   scale=1.5, thickness=2)

    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
